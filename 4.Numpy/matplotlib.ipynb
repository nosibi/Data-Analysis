# MatPlotLib 기본 그래프 그리기

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 기본 그래프

```python
x = np.arange(1,10)
y = x*5
```

```python
plt.plot(x,y)
plt.show()
```

## 색깔 바꾸기

```python
plt.plot(x,y,'r') # red : 'r' , blue : 'b' , black : 'k' , yellow : 'y' , green : 'g'
plt.show()
```

## 선 종류 변경하기

```python
plt.plot(x,y,'--')
plt.show()
```

```python
plt.plot(x,y,'or') # 빨간 점선, '종류색깔' 순으로 정의
plt.show()
```

```python
plt.plot(x,y,'ok')
plt.show()
```

## 라벨과 타이틀

```python
plt.plot(x,y)
plt.xlabel("xaxis") # x축 라벨명
plt.ylabel("yaxis") # y축 라벨명
plt.title("Basic Graph") # 타이틀
plt.show()
```

## 구간 확대/축소

```python
# x축을 2~3, y축은 5~20으로 확대
plt.plot(x,y)
plt.xlim(2,3)
plt.ylim(5,20)
plt.show()
```

## 레전드

여러개의 그래프를 그릴 시 구분을 위해 그래프마다 라벨을 작성. 이를 레전드라고 함.

```python
x = np.arange(1,10,0.1)
y1 = x * 0.2
y2 = np.sin(x)

plt.plot(x,y1,'b',label = "first")
plt.plot(x,y2,'r',label = "second")
plt.xlabel("axis X")
plt.ylabel("axis y")
plt.legend(loc = 'lower right') # loc 에 lower right, center right 등으로 라벨 위치 조정 가능
plt.show()
```

## 어노테이션

그래프에 화살표를 그린 후, 그 화살표에 문자열을 출력하는 기능

```python
x = np.arange(1,10)
y = x * 5
plt.plot(x,y,'b')
plt.annotate('critical point',xy = (2,10), xytext = (5,7),arrowprops={'color':'red'})
plt.show()
# annotate('제목',가리키는 지점, 설명할 지점, 화살표의 속성)
```

# 서브플롯

여러 그래프를 그려 위치를 지정

```python
# plt.subplot(row,column,location)
# location
# 1 2 
# 3 4
# 5 6

x = np.arange(1,10)
y1 = x * 5
y2 = x * 10

plt.subplot(2,1,1)
plt.plot(x,y1,'b',label = 'first')
plt.legend(loc = 'upper center')

plt.subplot(2,1,2)
plt.plot(x,y2,'r', label = 'second')
plt.legend(loc = 'upper center')

plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,'b',label = 'first')
plt.legend(loc = 'upper center')

plt.subplot(1,2,2)
plt.plot(x,y2,'r', label = 'second')
plt.legend(loc = 'upper center')

plt.show()
```

```python
# 2 x 2 형태
x = np.arange(1,10)
y1 = x * 5
y2 = x * 10
y3 = x * 15
y4 = x * 20

plt.subplot(2,2,1)
plt.plot(x,y1,'b',label = 'first')
plt.legend(loc = 'upper center')

plt.subplot(2,2,2)
plt.plot(x,y2,'r', label = 'second')
plt.legend(loc = 'upper center')

plt.subplot(2,2,3)
plt.plot(x,y3,'g',label = 'third')
plt.legend(loc = 'upper center')

plt.subplot(2,2,4)
plt.plot(x,y4,'k', label = 'fourth')
plt.legend(loc = 'upper center')
```
